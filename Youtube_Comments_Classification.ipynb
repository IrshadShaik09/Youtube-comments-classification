```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text = y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
            
    return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(ngram_range=(1,5))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB

def calculate_accuracy(X_tr,y_tr,X_te=X_test,y_te=y_test,clf=MultinomialNB(),vectorizer=vectorizer):
    
    X_train_vect = vectorizer.fit_transform(X_tr.TEXT.tolist())
    
    clf.fit(X=X_train_vect,y=y_tr)
    
    X_TEXT_vect = vectorizer.transform(X_te.TEXT.tolist())
    
    preds= clf.predict(X_TEXT_vect)
    
    return accuracy_score(preds,y_te)
```

```python
def labeling_rules_1(x):
    
    x=x.lower()
    
    rules =[
        "free" in x,
        "subs" in x,
        "http" in x
    ]
    
    if any (rules):
        return 1
    return -1
```

```python
def plot_accuracies(accs=accs):
    colors=list('rgbcmy')
    items_num = len(accs)
    cont = 1
    
    for x,y in accs.items():
        if x in ['true-label','random-labels','true-labes-best-clf']:
            plt.hlines(y,0,(items_num-2)*2, colors=colors.pop())
        else:
            plt.scatter(cont,y,s=100)
            cont+=2
    plt.legend(accs.keys(),loc = "center left", bbox_to_anchor =(1,0.5))
    plt.show()
    
plot_accuracies()
```

```python
def label_given_rules(df,rules_function,name,accs_dict=accs, verbose = True):
    # applying labeling rules to comments
    labels=[rules_function(label)for label in df.TEXT]
    #converting to numpy array
    labels = np.asarray(labels)
    
    initial_size = labels.shape[0]
    
    #trim points with no label
    X_train_al = df[labels != -1]
    labels = labels[labels != -1]
    
    final_size = labels.shape[0]
    
    acc = calculate_accuracy(X_train_al,labels)
    
    # print usefullinfo
    if verbose:
        print(f"Proportion of data points kept:{final_size/initial_size*100:.2f}%\n")
        print(f"{name} labeling has an accuracy of { acc*100:.2f}%\n")
        
    # saving accuracy to accuracies dict
    accs_dict[name]= acc
    
    return X_train_al,labels,acc
```

```python
def labeling_rules_2(x):
    x=x.lower()
    
    not_spam_rules = [
        "view" in x,
        "song" in x
    ]
    
    spam_rules =[
        "free" in x,
        "subs" in x,
        "gifts" in x,
        "follow" in x,
        "http" in x
    ]
    
    #classify depending on the rules
    
    if any(not_spam_rules):
        return 0
    
    if any(spam_rules):
        return 1
    return -1
```

```python
label_given_rules(X_train, labeling_rules_2,"second-iteration")
plot_accuracies()
```

```python
def labeling_rules_3(x):
    x=x.lower()
    
    not_spam_rules = [
        "view" in x,
        "song" in x,
        len(x)< 30
    ]
    #defineing list of rules to classify as spam
    spam_rules =[
        "free" in x,
        "subs" in x,
        "gifts" in x,
        "follow" in x,
        "http" in x,
        "check out" in x
    ]
    
    #classify depending on the rules
    
    if any(not_spam_rules):
        return 0
    
    if any(spam_rules):
        return 1
    return -1
```
